### Два интерфейса

В библиотеке реализована два способа взаимодействия - MATLAB-подобный и объектно-ориентированный

В MatLab подобном интерфейсе сначала создаётся "холст"(фигура) для графика, а затем путём переключения между установленными частями фигуры можно выводить на текущий выбранный график. Т.е. данный интерфейс запоминает, в какой части вы находитесь и рисует только там

```python
plt.figure() # Создаем рисунок для графика
# Создаем первую из двух областей графика и задаем текущую ось
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# Создаем вторую область и задаем текущую ось
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

В объектно-ориентированном подходе интерфейс функций рисования является методами явным образом определяемых объектов Figure и Axes

```python
# Сначала создаем сетку графиков
# ax будет массивом из двух объектов Axes
fig, ax = plt.subplots(2)
# Вызываем метод plot() соответствующего объекта
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

### Цвета и стили линий

Цвет графика задаётся аргументом color

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x - 0), color='blue') # Задаем цвет по названию
ax.plot(x, np.sin(x - 1), color='g') # Краткий код цвета (rgbcmyk)
ax.plot(x, np.sin(x - 2), color='0.75') # Шкала оттенков серого цвета, значения в диапазоне от 0 до 1
ax.plot(x, np.sin(x - 3), color='#FFDD44') # 16-ричный код (RRGGBB от 00 до FF)
ax.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # Кортеж RGB, значения 0 и 1
ax.plot(x, np.sin(x - 5), color='chartreuse'); # Поддерживаются все названия цветов HTML
```

```python
plt.plot(x, np.sin(x - 0), color='blue') # Задаем цвет по названию
plt.plot(x, np.sin(x - 1), color='g') # Краткий код цвета (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Шкала оттенков серого цвета, значения в диапазоне от 0 до 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # 16-ричный код (RRGGBB от 00 до FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # Кортеж RGB, значения 0 и 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # Поддерживаются все названия цветов HTML
```

Стиль линий настраивается с помощью ключевого слова linestyle

```python
plt.plot(x, x - 1, linestyle='-') # сплошная линия 'solid'
plt.plot(x, x - 2, linestyle='--') #штриховая линия 'dashed'
plt.plot(x, x - 3, linestyle=':') # пунктирная 'dotted'
plt.plot(x, x - 4, linestyle='-.') # штрихпунктирная 'dashdot'
```

Можно задавать стиль и цвет одним аргументом

```python
plt.plot(x, x + 0, '-g') # сплошная линия зеленого цвета
plt.plot(x, x + 1, '--c') # штриховая линия голубого цвета
plt.plot(x, x + 2, '-.k') # штрихпунктирная линия черного цвета
plt.plot(x, x + 3, ':r'); # пунктирная линия красного цвета
```

## Простые диаграммы рассеяния

### Построение диаграмм рассеяния с помощью функции plot

Диаграммы рассеивания - набор несоединённых точек на плоскости. Для построения диаграммы рассеивания достаточно указать макер, которым будут обозначаться точки вместо обозначнения стиля линии

```python
for marker in ['o', '.', ',', 'x', '+']:
    plt.plot(
        np.random.rand(10), 
        np.random.rand(10), 
        marker,
        label=f"marker={marker}")
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

Если совместить обозначние маркера и стиля линии, то можно получить график с выделенными точками, соединёнными прямыми линиями 

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, np.sin(x), 'o-')
```

```python
for marker in ['o', '.', ',', 'x', '+']:
    plt.plot(
        np.random.rand(10), 
        np.random.rand(10), 
        marker + '-',
        label=f"marker={marker}"
        )
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

С помощью дополнительных ключевых аргументов функции plt.plot можно задавать множество свойств линий и маркеров

```python
plt.plot(x, y, '-h', color='magenta', markersize=12, linewidth=6, markerfacecolor='grey', 
        markeredgecolor='blue', markeredgewidth=3)
```

### Построение диаграмм рассеяния с помощью функции plt.scatter

```python
plt.scatter(x, y, marker='o')
```

Основное различие между функциями plt.scatter и plt.plot состоит в том, что с помощью первой можно создавать диаграммы рассеяния с индивидуально задаваемыми (или выбираемыми в соответствии с данными) свойствами каждой точки (размер, цвет заливки, цвет рамки и т. д.)

alpha позволяет настроить уровень прозрачности 

```python
rng = np.random
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000 * rng.randn(100)
```

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

scatter позволяет настраивать каждую точку в отдельности, поэтому на больших данных она может работать гораздо дольше чем plot

## Визуализация погрешностей

Простые планки погрешностей можно создать с помощью вызова всего одной функции

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * rng.rand(50)
```

```python
plt.errorbar(x, y, yerr=dy, fmt=':b')
plt.plot(x, np.sin(x))
```

fmt - это код форматирования, управлябщий внешним видом линий и точек (считай цвета и стили обычного plot)

Функция имеет множество настроек

```python
plt.errorbar(x, y, yerr=dy, fmt='.', color='black',
            ecolor='lightgrey', elinewidth=3)
```
